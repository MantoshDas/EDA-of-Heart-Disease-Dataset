# c. Analyze Distribution of Heart Disease (Target Variable):
 --> Visualize the distribution of heart disease cases using a bar chart or pie chart.

```python
heart_disease_counts = df_clean['HeartDiseaseorAttack'].value_counts()
colors = ['blue', 'red']
plt.figure(figsize=(6, 4))
sns.barplot(x=heart_disease_counts.index, y=heart_disease_counts.values, palette=colors)
plt.xlabel('Heart Disease or Attack')
plt.ylabel('Count')
plt.title('Distribution of Heart Disease Cases')
plt.xticks(range(len(heart_disease_counts.index)), ['No Disease', 'Disease'])
plt.show()
```

# b. Visualize Correlations Between Variables:
--> Use scatter plots or correlation matrices to explore relationships between continuous variables.

```python
corr = df_clean[['BMI', 'MentHlth', 'PhysHlth', 'Age']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# c. Compare Heart Disease Across Demographic Groups:
--> Analyze the prevalence of heart disease across different demographic groups (e.g., age, gender, education) using bar charts or pivot tables.

# For Age

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='HeartDiseaseorAttack', data=df_clean, palette='viridis')
plt.title('Heart Disease Prevalence Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Prevalence of Heart Disease')
plt.xticks(rotation=45)
plt.show()
```

# For Sex

```python
plt.figure(figsize=(10, 6))
color = ['blue', 'green']
sns.barplot(x='Sex', y='HeartDiseaseorAttack', data=df_clean, palette=color)
plt.title('Heart Disease Prevalence Across Sex Groups')
plt.xlabel('Sex')
plt.ylabel('Prevalence of Heart Disease')
plt.xticks(rotation=45)
plt.show()
```

# For Education

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Education', y='HeartDiseaseorAttack', data=df_clean, palette='rocket')
plt.title('Heart Disease Prevalence Across Education')
plt.xlabel('Education')
plt.ylabel('Prevalence of Heart Disease')
plt.xticks(rotation=45)
plt.show()
```

```python
def run_eda(file_path):
    df = pd.read_csv(file_path)
    summary = {
        "rows": len(df),
        "columns": len(df.columns),
        "missing_values": df.isnull().sum().sum(),
    }

    if "HeartDiseaseorAttack" in df.columns:
        summary["correlation_with_target"] = (
            df.corr(numeric_only=True)["HeartDiseaseorAttack"].to_dict()
        )
    else:
        summary["correlation_with_target"] = {}

    return summary
```

```python
result = run_eda("/content/heart_disease_health_indicators_BRFSS2015 2 (1).csv")
print(result)
```
